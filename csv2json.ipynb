# Converts a .csv to .json file

```python
import json
import csv
```

```python
data = csv.reader(open('data.csv', newline=''))
cars = {}
for row in data:
    model_name = row[2]
    if model_name not in cars:
        cars[model_name] = {}
    cars[model_name]['brand'] = row[1]
    if row[0] == 'Standard Service':
        if row[3] != '0':
            cars[model_name]['standard-petrol'] = row[-1]
        if row[4] != '0':
            cars[model_name]['standard-diesel'] = row[-1]
        if row[5] != '0':
            cars[model_name]['standard-cng'] = row[-1]
    elif row[0] == 'Comprehensive Service':
        if row[3] != '0':
            cars[model_name]['comprehensive-petrol'] = row[-1]
        if row[4] != '0':
            cars[model_name]['comprehensive-diesel'] = row[-1]
        if row[5] != '0':
            cars[model_name]['comprehensive-cng'] = row[-1]
    elif row[0] == 'Basic Service':
        if row[3] != '0':
            cars[model_name]['basic-petrol'] = row[-1]
        if row[4] != '0':
            cars[model_name]['basic-diesel'] = row[-1]
        if row[5] != '0':
            cars[model_name]['basic-cng'] = row[-1]
```

```python
json.dump(cars, open('cars.json', 'w'), indent=2)
print('Wrote data.csv to cars.json')
```
