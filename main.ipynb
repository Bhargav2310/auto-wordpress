```python
cars = json.load(open('cars.json', 'r'))
```

```python
brand_xpaths = {
    'Chevrolet': {
        'id': 'brand-14',
        'xpath': '//*[@id="brand-14"]/label'
    }, 
    'Daewoo': {
        'id': 'brand-15',
        'xpath': '//*[@id="brand-15"]/label'
    },
    'Datsun': {
        'id': 'brand-16',
        'xpath': '//*[@id="brand-16"]/label'
    },
    'Fiat': {
        'id': 'brand-20',
        'xpath': '//*[@id="brand-20"]/label'
    },
    'Ford': {
        'id': 'brand-21',
        'xpath': '//*[@id="brand-21"]/label'
    },
    'Honda': {
        'id': 'brand-22',
        'xpath': '//*[@id="brand-22"]/label'
    },
    'Hyundai': {
        'id': 'brand-23',
        'xpath': '//*[@id="brand-23"]/label'
    },
    'Mahindra': {
        'id': 'brand-24',
        'xpath': '//*[@id="brand-24"]/label'
    },
    'Maruti Suzuki': {
        'id': 'brand-25',
        'xpath': '//*[@id="brand-25"]/label'
    },
    'Mitsubishi': {
        'id': 'brand-26',
        'xpath': '//*[@id="brand-26"]/label'
    },
    'Nissan': {
        'id': 'brand-28',
        'xpath': '//*[@id="brand-28"]/label'
    },
    'Renault': {
        'id': 'brand-29',
        'xpath': '//*[@id="brand-29"]/label'
    },
    'Skoda': {
        'id': 'brand-30',
        'xpath': '//*[@id="brand-30"]/label'
    },
    'TATA': {
        'id': 'brand-31',
        'xpath': '//*[@id="brand-31"]/label'
    },
    'Toyota': {
        'id': 'brand-32',
        'xpath': '//*[@id="brand-32"]/label'
    },
    'Volkswagen': {
        'id': 'brand-33',
        'xpath': '//*[@id="brand-33"]/label'
    },
}
```

```python
i = 1
for model in cars:
    driver.get('http://103.250.185.39/~ioitcollege/corifycar/wp-admin/post-new.php?post_type=services')
    brand = cars[model]['brand']
    log(f'{i}. GET {brand} {model} ... {round(i/len(cars) * 100, 2)}%')
    title = driver.find_element_by_xpath('//*[@id="title"]')
    title.send_keys(model)

    # DATA
    if 'basic-petrol' in cars[model]:
        driver.find_element_by_xpath('//*[@id="acf-field_60323ac4aba50-field_60324053cba63"]').send_keys(cars[model]['basic-petrol'])
        log(f'is-basic-petrol: True   Price: {cars[model]["basic-petrol"]}')
    else:
        log('is-basic-petrol: False')

    if 'basic-diesel' in cars[model]:
        driver.find_element_by_xpath('//*[@id="acf-field_60323ac4aba50-field_6032406acba64"]').send_keys(cars[model]['basic-diesel'])
        log(f'is-basic-diesel: True   Price: {cars[model]["basic-diesel"]}')
    else:
        log('is-basic-diesel: False')

    if 'basic-cng' in cars[model]:
        driver.find_element_by_xpath('//*[@id="acf-field_60323ac4aba50-field_6032407bcba65"]').send_keys(cars[model]['basic-cng'])
        log(f'is-basic-cng: True   Price: {cars[model]["basic-cng"]}')
    else:
        log('is-basic-cng: False')

    if 'standard-petrol' in cars[model]:
        driver.find_element_by_xpath('//*[@id="acf-field_603240f4c5ace-field_603240f4c5acf"]').send_keys(cars[model]['standard-petrol'])
        log(f'is-standard-petrol: True   Price: {cars[model]["standard-petrol"]}')
    else:
        log('is-standard-petrol: False')

    if 'standard-diesel' in cars[model]:
        driver.find_element_by_xpath('//*[@id="acf-field_603240f4c5ace-field_603240f4c5ad0"]').send_keys(cars[model]['standard-diesel'])
        log(f'is-standard-diesel: True   Price: {cars[model]["standard-diesel"]}')
    else:
        log('is-standard-diesel: False')

    if 'standard-cng' in cars[model]:
        driver.find_element_by_xpath('//*[@id="acf-field_603240f4c5ace-field_603240f4c5ad1"]').send_keys(cars[model]['standard-cng'])
        log(f'is-standard-cng: True   Price: {cars[model]["standard-cng"]}')
    else:
        log('is-standard-cng: False')

    if 'comprehensive-petrol' in cars[model]:
        driver.find_element_by_xpath('//*[@id="acf-field_6032411752da3-field_6032411752da4"]').send_keys(cars[model]['comprehensive-petrol'])
        log(f'is-comprehensive-petrol: True   Price: {cars[model]["comprehensive-petrol"]}')
    else:
        log('is-comprehensive-petrol: False')

    if 'comprehensive-diesel' in cars[model]:
        driver.find_element_by_xpath('//*[@id="acf-field_6032411752da3-field_6032411752da5"]').send_keys(cars[model]['comprehensive-diesel'])
        log(f'is-comprehensive-diesel: True   Price: {cars[model]["comprehensive-diesel"]}')
    else:
        log('is-comprehensive-diesel: False')

    if 'comprehensive-cng' in cars[model]:
        driver.find_element_by_xpath('//*[@id="acf-field_6032411752da3-field_6032411752da6"]').send_keys(cars[model]['comprehensive-cng'])
        log(f'is-comprehensive-cng: True   Price: {cars[model]["comprehensive-cng"]}')
    else:
        log('is-comprehensive-cng: False')

    # SERVICE TYPE
    service_div = driver.find_element_by_xpath('//*[@id="service-typediv"]')
    driver.execute_script("arguments[0].scrollIntoView();", service_div)
    standard, comprehensive, basic, diesel, petrol, cng = (False for _ in range(6))
    for prop in cars[model]:
        prop: str = prop
        if prop.__contains__('standard'):
            checkbox = driver.find_element_by_xpath('//*[@id="service-type-17"]/label')
            if not checkbox.is_selected() and not standard:
                standard = True
                driver.find_element_by_xpath('//*[@id="service-type-17"]/label').click()
        if prop.__contains__('comprehensive'):
            checkbox = driver.find_element_by_xpath('//*[@id="service-type-18"]/label')
            if not checkbox.is_selected() and not comprehensive:
                comprehensive = True
                driver.find_element_by_xpath('//*[@id="service-type-18"]/label').click()
        if prop.__contains__('basic'):
            checkbox = driver.find_element_by_xpath('//*[@id="service-type-19"]/label')
            if not checkbox.is_selected() and not basic:
                basic = True
                driver.find_element_by_xpath('//*[@id="service-type-19"]/label').click()

        if prop.__contains__('diesel'):
            checkbox = driver.find_element_by_xpath('//*[@id="fuel-type-12"]/label')
            if not checkbox.is_selected() and not diesel:
                diesel = True
                driver.find_element_by_xpath('//*[@id="fuel-type-12"]/label').click()
        if prop.__contains__('petrol'):
            checkbox = driver.find_element_by_xpath('//*[@id="fuel-type-11"]/label')
            if not checkbox.is_selected() and not petrol:
                petrol = True
                driver.find_element_by_xpath('//*[@id="fuel-type-11"]/label').click()
        if prop.__contains__('cng'):
            checkbox = driver.find_element_by_xpath('//*[@id="fuel-type-13"]/label')
            if not checkbox.is_selected() and not cng:
                cng = True
                driver.find_element_by_xpath('//*[@id="fuel-type-13"]/label').click()

    # BRAND
    driver.find_element_by_xpath(brand_xpaths[cars[model]['brand']]['xpath']).click()

    while True:
        autosave = driver.find_element_by_xpath('//*[@id="post-status-info"]/tbody/tr/td[2]/span').text
        if autosave.__contains__(':') or autosave.__contains__('saved'):
            break

    p = driver.find_element_by_xpath('//*[@id="submitdiv"]')
    driver.execute_script("arguments[0].scrollIntoView();", p)
    driver.find_element_by_xpath('//*[@id="publish"]').click()

    log ('Publishing ...')
    while True:
        try:
            span = driver.find_element_by_xpath('//*[@id="post-status-display"]').text
            if span == 'Published':
                log('Published\n')
                break
        except:
            pass
    i += 1
```
